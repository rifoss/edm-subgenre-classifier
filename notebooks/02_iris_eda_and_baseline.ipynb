```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the data, using as_frame=True to keep it in a DataFrame structure
iris = load_iris(as_frame=True)

# Combine features and target into a single DataFrame for easier plotting
iris_df = iris.frame
```

```python
# Check for missing values and data types
print(iris_df.info())
```

```python
print(iris_df.describe())
```

```python
iris_df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# We rename the target column for plotting clarity (0, 1, 2 --> setosa, versicolor, virginica)
iris_df['species'] = iris.target_names[iris.target]

sns.pairplot(iris_df, hue='species', height=2.5, diag_kind='kde')
plt.suptitle("Pairwise Feature Distribution Colored by Species", y=1.02)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

x, y = iris.data, iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_model.fit(x_train, y_train)

print("K-Nearest Neighbors baseline model trained.")
```

```python
baseline_accuracy = knn_model.score(x_test, y_test)
print(f"KNN Baseline Accuracy (K=3): {baseline_accuracy:.4f}")
```
